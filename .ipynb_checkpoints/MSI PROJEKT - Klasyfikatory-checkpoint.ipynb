# Klasyfikatory

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
from sklearn import model_selection, neighbors, metrics

from sklearn.ensemble import RandomForestClassifier, ExtraTreesClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.linear_model import SGDClassifier

from IPython.display import HTML, display

import time
import random
```

```python
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/breast-cancer-wisconsin/wdbc.data'
names = ['id_number', 'diagnosis', 'radius_mean',
         'texture_mean', 'perimeter_mean', 'area_mean',
         'smoothness_mean', 'compactness_mean', 'concavity_mean',
         'concave_points_mean', 'symmetry_mean',
         'fractal_dimension_mean', 'radius_se', 'texture_se',
         'perimeter_se', 'area_se', 'smoothness_se',
         'compactness_se', 'concavity_se', 'concave_points_se',
         'symmetry_se', 'fractal_dimension_se',
         'radius_worst', 'texture_worst', 'perimeter_worst',
         'area_worst', 'smoothness_worst',
         'compactness_worst', 'concavity_worst',
         'concave_points_worst', 'symmetry_worst',
         'fractal_dimension_worst']
df = pd.read_csv(url, names=names)

df.set_index(['id_number'], inplace = True)
```

```python
X = df.iloc[:, df.columns != 'diagnosis']
y = df.iloc[:, df.columns == 'diagnosis']

X_train, X_test, y_train, y_test = model_selection.train_test_split(X, y, test_size = 0.70, random_state=0)

clfs = [
            [
                'RandomForestClassifier',
                RandomForestClassifier(random_state=0),
                {
                    'n_estimators': np.arange(5,10),
                    'max_depth': [2, 3, 4],
                    'bootstrap': [True, False],
                    'max_features': ['auto', 'sqrt', 'log2', None],
                    'criterion': ['gini', 'entropy']
                }
            ],
            [
                'ExtraTreesClassifier',
                ExtraTreesClassifier(random_state=0),
                {
                    'n_estimators': np.arange(5,10),
                    'max_depth': [2, 3, 4],
                    'bootstrap': [True, False],
                    'max_features': ['auto', 'sqrt', 'log2', None],
                    'criterion': ['gini', 'entropy']
                }
            ],
            [
                'DecisionTreeClassifier',
                DecisionTreeClassifier(random_state=0),
                {
                    'max_depth': [2, 3, 4],
                    'max_features': ['auto', 'sqrt', 'log2', None],
                }
            ],
            [
                'SVC',
                SVC(probability = True),
                {
                    'C': [1, 10, 100, 1000],
                    'gamma': [0.001, 0.0001],
                    'kernel': ['rbf'],
                    'class_weight':['balanced', None],
                    'probability' : [True, False]
                 }
            ],
            [
                'KNeighborsClassifier',
                KNeighborsClassifier(),
                {
#                     'n_neighbors': np.arange(5,10),
#                     'weights': ['uniform', 'distance'],
#                     'algorithm': ['auto', 'ball_tree', 'kd_tree', 'brute']
                }
            ],
            [
                'GaussianNB',
                GaussianNB(priors=None),
                {
                }
            ]
        ]
```

```python
y_train = y_train.values.ravel() 
y_test = y_test.values.ravel()
```

## Określenie testowych hiperparametrów

```python
best_params = []
for name, clf, params in clfs:
    np.random.seed(0)
    start = time.time()
    param_dist = params
    cv= model_selection.GridSearchCV(clf, cv = 10,param_grid=param_dist, n_jobs = 3)
    cv.fit(X_train, y_train)
    best_params.append(cv.best_params_)
    print('Najlepsze parametry klasyfikatora %s:' % name, cv.best_params_)
    end = time.time()
```

```python
clfs[0][1].set_params(
                n_estimators = best_params[0]['n_estimators'],
                max_depth = best_params[0]['max_depth'],
                max_features = best_params[0]['max_features'],
                criterion = best_params[0]['criterion']
              )

clfs[1][1].set_params(
                n_estimators = best_params[1]['n_estimators'],
                max_depth = best_params[1]['max_depth'],
                max_features = best_params[1]['max_features'],
                criterion = best_params[1]['criterion']
              )

clfs[2][1].set_params(
                max_depth = best_params[2]['max_depth'],
                max_features = best_params[2]['max_features'],
              )

clfs[3][1].set_params(
                C = best_params[3]['C'],
                gamma = best_params[3]['gamma'],
                kernel = best_params[3]['kernel'],
                class_weight = best_params[3]['class_weight'],
                probability = best_params[3]['probability']
              )
```

```python
for name, clf, params in clfs:
    clf.fit(X_train, y_train)
    print(clf)
```

## Wybór najistotniejszych parametrów

```python
importances = clfs[0][1].feature_importances_
indices = np.argsort(importances)[::-1]
names_index = names[2:]

index = np.arange(len(names_index))

importance_desc = sorted(importances) 
feature_space = []
for i in range(29, -1, -1):
    feature_space.append(names_index[indices[i]])

fig, ax = plt.subplots(figsize=(12, 12))

plt.title('Ważność cech dla klasyfikatora Random Forest Model')
plt.barh(index, importance_desc, align="center", color = 'Blue')
plt.yticks(index, feature_space)
plt.ylim(-1, 30)
plt.xlim(0, max(importance_desc))
plt.xlabel('Średnia znaczenia')
plt.ylabel('Cecha')

plt.show()
plt.close()
```

## Zastosowane metryki oceny jakości:

### 1) Macierz pomyłek:

```python
accuracies = []
error_rates = []
predictions = []
for name, clf, params in clfs:
    prediction = clf.predict(X_test)
    predictions.append(prediction)
    test_crosstb = pd.crosstab(index = y_test, columns = prediction)
    test_crosstb = test_crosstb.rename(columns= {'B': 'Benign', 'M': 'Malignant'})
    test_crosstb.index = ['Benign', 'Malignant']
    test_crosstb.columns.name = ''
    
    accuracy = clf.score(X_test, y_test)
    accuracies.append(accuracy)
    
    error_rate = 1 - accuracy
    error_rates.append(error_rate)
    
    print("%s:" % name)
    print(test_crosstb)
    print("\n")
    print("Średnia dokładność: %.4f" % accuracy)
    print("Stopa błędu: %.4f\n" % error_rate)
```

### 2) k-foldowa walidacja krzyżowa:

```python
cv_scores = []
for name, clf, params in clfs:
    score = model_selection.cross_val_score(clf, X_train, y_train, cv = 10)
    cv_scores.append([score.mean(), score.std()])
    print("Dokladnosc klasyfikatora %s: %.4f (+/- %.4f)" % (name, score.mean(), score.std()))
```

### 3) wykres krzywej ROC:

```python
aucs = []
fig, ax = plt.subplots(figsize=(10, 10))

for name, clf, params in clfs:
    predict_proba = clf.predict_proba(X_test)[:, 1]
    fpr, tpr, _ = metrics.roc_curve(y_test == 'M', predict_proba)
    auc = metrics.auc(fpr, tpr)
    aucs.append(auc)
    plt.plot(fpr, tpr, linewidth = 3, label=name)
    print("AUC dla %s: %.4f" % (name, auc))
    
    
plt.title('Krzywa ROC'.format(auc))

plt.plot([0, 1], [0, 1], 'k--', lw = 2)
plt.plot([0, 0], [1, 0], 'k--', lw = 2, color = 'Black')
plt.plot([1, 0], [1, 1], 'k--', lw = 2, color = 'Black')
plt.xlabel('FPR')
plt.ylabel('TPR')
plt.legend(loc="lower right")
plt.show()
plt.close()
```

### 4) precyzja(precision) i pełność(recall)

```python
# Precision: also the positive predictive value, is the number of correct predictions divided by the number of correct predictions plus false positives, so $tp / (tp + fp)$
# Recall: also known as the sensitivity, is the number of correct predictions divided by the total number of instances so $tp / (tp + fn)$ where $fn$ is the number of false negatives
# f1-score: this is defined as the weighted harmonic mean of both the precision and recall, where the f1-score at 1 is the best value and worst value at 0, as defined by the documentation
# support: number of instances that are the correct target values
precisions = []
recalls = []
for prediction in predictions:
    precision, recall, fscore, support = metrics.precision_recall_fscore_support(prediction, y_test, average='macro')
    precisions.append(precision)
    recalls.append(recall)
```

```python
data = []
for clf, accuracy, er, score, auc, precision, recall in zip(clfs, accuracies, error_rates, cv_scores, aucs, precisions, recalls):
    data.append([clf[0], "%.4f" % accuracy, "%.4f" % er, "%.4f (+/- %.4f)" % (score[0], score[1]), "%.4f" % auc, "%.4f" % precision, "%.4f" % recall])

display(HTML(
    '<table><tr>{}</tr></table>'.format(
        '</tr><tr>'.join(
            '<td>{}</td>'.format('</td><td>'.join(str(_) for _ in row)) for row in data)
        )
 ))
```
